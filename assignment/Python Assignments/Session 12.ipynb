```python
# 1Write a lambda function that takes a price in rupees and returns the price after adding 18% GST.
# Test it on the prices 100, 250, and 500.

gst = lambda price: price + price*18/100

print(gst(100))
print(gst(250))
print(gst(500))

```

```python
# 2Given a list of song titles from Spotify with extra spaces and inconsistent casing, use map() and a lambda function to clean each title so 
# that it is stripped of spaces and converted to title case (e.g., ' shape OF you ' → 'Shape Of You').

songs = [
    " daku BY Sidhu ",
    " blinding LIGHTS ",
    " belIEver ",
    "   senorita   ",
    " levitating "
]

clean_list= list(map(lambda song:song.strip().title(),songs))

print(clean_list)
```

```python
# Use filter() and a lambda function to extract only those Flipkart product names from a list that start with the letter 'S'
# (case-insensitive).

Flipkart_product_names=["Shose","tshart","shrt","jins","Short"]

flitar_list=list(filter(lambda item :item.lower().startswith("s"),Flipkart_product_names))

print(flitar_list)
```

```python
from functools import reduce
4.# Given a list of order amounts from a Zomato cart [120, 340, 560, 80], use reduce() from functools to calculate the total bill amount.
cart=[120, 340, 560, 80]

total= reduce(lambda a,b:a+b , cart)

print("TOTAL AMOUNT OF CART IS ", total)
```

```python
# 5Use ChatGPT or Copilot to generate a Python code snippet that uses map(), filter(), and reduce() together to process a list of numbers:
# first double each number, then filter to keep only numbers greater than 100, and finally sum the result. Paste and test the generated 
# code with the list [40, 60, 80, 120].
from functools import reduce

numbers = [40, 60, 80, 120]

# Step 1: Double each number
doubled = list(map(lambda x: x * 2, numbers))
print("After map():", doubled)

# Step 2: Keep only numbers greater than 100
filtered = list(filter(lambda x: x > 100, doubled))
print("After filter():", filtered)

# Step 3: Sum all remaining numbers
total = reduce(lambda a, b: a + b, filtered)
print("Final Sum:", total)
```
