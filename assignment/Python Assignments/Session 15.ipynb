```python
class InvalidDurationError(Exception):
    pass


def Sum_of_playlist(songs):
    Total_second = 0

    for time in songs:
        if time < 0:
            raise InvalidDurationError("Song time cannot be negative")
        Total_second += time

    total_minutes = Total_second / 60
    return total_minutes


songs = [60, 50, -120, 80]

try:
    result = Sum_of_playlist(songs)
    print("Total playlist duration:", result, "minutes")

except InvalidDurationError as e:
    print("Error:", e)
```

```python
# Build a Flipkart-style order summary: ask the user for item price and quantity, then calculate and print total price. 
# Use try-except-else-finally blocks to handle ValueError for invalid input, print the total if successful, and always print 
# 'Thank you for shopping!' in the finally block.


try:
    price=int(input("Enter a Price of Product"))
    quantity=int(input("Enter a num of Quantity"))
    total_price=price*quantity
except ValueError:
    print("invelead Input")
else:
    print("total price is :",total_price)
finally:
    print("Thank you for shopping!")
    

```
