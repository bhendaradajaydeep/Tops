```python
# 2. Write a function update_playlist_price(playlist, new_price) that updates the price of a given playlist in the playlist_prices dictionary. 
# it by updating the price of any one playlist and printing the updated dictionary.

playlist_prices={"hindi song":2000,"panjabi _song":2100,"gujrati song":1000,"bhojpuri sing":900,"kumar sanu song":3000}

def update_playlist_price(playlist, new_price):
    playlist_prices[playlist]=new_price

playlist_name=str(input("Enter a pley list name:"))
playlist_new_prices=int(input("Enter a list new price:"))
update_playlist_price(playlist_name,playlist_new_prices)

print(playlist_prices)
    

```
