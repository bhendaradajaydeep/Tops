```python
# Create a Python program that takes the number of followers as input and uses if, elif, and else to print 'Micro Influencer'
# if followers < 10,000,'Rising Star' if between 10,000 and 100,000, and 'Celebrity' if above 100,000.


name=str(input("Enter your user name: "))
followers=int(input("Enter your followers "))

if followers < 10000:
    print(f"{name} is a Micro Influencer")
elif followers > 10000 and followers <100000:
    print(f"{name} is a Rising Star")
else :
    print(f" {name} is a Celebrity ")

```

```python
4.
# Write a Python program using nested if statements: take a user's entered Flipkart cart value and payment method ('UPI', 'Card', 'Cash').
# If the cart value is above 1000 and payment method is 'UPI', print 'Eligible for 10% cashback'; if above 1000 and payment is not 'UPI',
# print 'Eligible for 5% cashback'; else print 'No cashback'.

value = int(input("Enter Flipkart cart value: "))
payment_method = input("Enter payment method (UPI/Card/Cash): ").upper()

if value > 1000 and payment_method=="UPI":
    if payment_method == "UPI":
        print("Eligible for 10% cashback.  your final amount is: ", value - value*10/100 )
elif value >1000:
    print("Eligible for 5% cashback", value - value*5/100)
else:
    print("No cashback")
```
