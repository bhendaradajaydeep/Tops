```python
import pandas as pd
import numpy as np

# Create sample dataset
np.random.seed(10)

cities = ["Ahmedabad", "Surat", "Vadodara"]

orders = pd.DataFrame({
    "Order_ID": range(1, 1001),
    "City": np.random.choice(cities, 1000, p=[0.5, 0.3, 0.2])
})

sample_size = 60

# Empty list for samples
samples = []

# Count orders by city
city_counts = orders["City"].value_counts()

for city in city_counts.index:
    proportion = city_counts[city] / len(orders)
    n = round(proportion * sample_size)

    city_sample = orders[orders["City"] == city].sample(n=n, random_state=1)
    samples.append(city_sample)

# Combine all samples
stratified_sample = pd.concat(samples)

print(stratified_sample)
print("\nTotal Sample Size:", len(stratified_sample))
```

```python
def systematic_sampling(reviews, interval):
    sample = []

    for i in range(0, len(reviews), interval):
        sample.append(reviews[i])

    return sample


# Create 500 reviews
reviews = [f"Review {i}" for i in range(1, 501)]

sample_reviews = systematic_sampling(reviews, 10)

print("Number of Sampled Reviews:", len(sample_reviews))
print(sample_reviews)
```

```python
import pandas as pd
import numpy as np

np.random.seed(20)

teams = [
    "CSK",
    "MI",
    "RCB",
    "KKR",
    "GT",
    "RR",
    "SRH",
    "DC"
]

# Create dataset of 400 fans
fans = pd.DataFrame({
    "Fan_ID": range(1, 401),
    "Favorite_Team": np.random.choice(
        teams,
        400,
        p=[0.18, 0.16, 0.15, 0.12, 0.10, 0.10, 0.10, 0.09]
    )
})

sample_size = 40

result = []

for team in teams:
    group = fans[fans["Favorite_Team"] == team]

    proportion = len(group) / len(fans)

    n = max(1, round(proportion * sample_size))

    result.append(group.sample(n=n, random_state=10))

final_sample = pd.concat(result)

print(final_sample)

print("\nTotal Sample Size:", len(final_sample))
print("\nFans by Team:")
print(final_sample["Favorite_Team"].value_counts())
```
