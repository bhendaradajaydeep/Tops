```python
# ============================================
# SESSION 6 - Probability Basics
# ============================================

# --------------------------------------------
# Task 1
# Sample Space
# Rolling a Die and Flipping a Coin
# --------------------------------------------

print("========== Task 1 ==========")

die = [1, 2, 3, 4, 5, 6]
coin = ["Head", "Tail"]

sample_space = []

for d in die:
    for c in coin:
        sample_space.append((d, c))

print("Sample Space:")
print(sample_space)
print("Total Outcomes =", len(sample_space))


# --------------------------------------------
# Task 2
# Classical Probability
# Drawing an Even Card
# --------------------------------------------

print("\n========== Task 2 ==========")

cards = list(range(1, 11))

even_cards = [card for card in cards if card % 2 == 0]

probability = len(even_cards) / len(cards)

print("Cards =", cards)
print("Even Cards =", even_cards)
print("Favorable Outcomes =", len(even_cards))
print("Total Outcomes =", len(cards))
print("Probability =", probability)


# --------------------------------------------
# Task 3
# Spotify Playlist
# Probability of NOT Arijit Singh Song
# --------------------------------------------

print("\n========== Task 3 ==========")

total_songs = 8
arijit_songs = 3

not_arijit = total_songs - arijit_songs

probability = not_arijit / total_songs

print("Total Songs =", total_songs)
print("Arijit Songs =", arijit_songs)
print("Non-Arijit Songs =", not_arijit)
print("Probability =", probability)


# --------------------------------------------
# Task 4
# Addition Rule
# Pizza OR Burger
# --------------------------------------------

print("\n========== Task 4 ==========")

pizza = 0.3
burger = 0.4
both = 0.1

probability = pizza + burger - both

print("P(Pizza) =", pizza)
print("P(Burger) =", burger)
print("P(Both) =", both)
print("P(Pizza OR Burger) =", probability)


# --------------------------------------------
# Task 5
# Conditional Probability Explanation
# --------------------------------------------

print("\n========== Task 5 ==========")

phone_probability = 0.2
mobile_interest_probability = 0.5

print("Probability of adding phone to cart =", phone_probability)
print("Probability if user is interested in mobiles =", mobile_interest_probability)

print("\nExplanation:")
print("""
Conditional probability means that the chance of an event changes
when additional information is available.

In this Flipkart example:

- Normally, there is only a 20% chance (0.2) that a user adds a phone to the cart.
- If we already know the user is interested in mobiles,
  the probability increases to 50% (0.5).

This shows that knowing a user's interest helps predict
their buying behavior more accurately.
""")
```
