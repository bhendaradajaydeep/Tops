```python

import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson

p = 0.85      
trials = 20

results = []

for i in range(trials):
    login = 1 if random.random() < p else 0
    results.append(login)

print("Login Results:")
print(results)

print("Successful Logins:", sum(results))
print("Failed Logins:", trials - sum(results))
```

```python
# ========== Task 2 ==========


def upi_payment_probability(n, k, p):
    return binom.pmf(k, n, p)

probability = upi_payment_probability(
    n=5,
    k=3,
    p=0.9
)

print("Probability of exactly 3 successful payments =", probability)
```

```python
average_playlists = 4
exact_playlists = 6

probability = poisson.pmf(
    exact_playlists,
    average_playlists
)

print("Probability of exactly 6 playlists =", probability)
```

```python
# ========= Task 4 =========="

average_reviews = 10

reviews = np.random.poisson(
    lam=average_reviews,
    size=12
)

print("Reviews received in 12 random hours:")
print(reviews)

hours = np.arange(1, 13)

plt.figure(figsize=(8,5))
plt.bar(hours, reviews)

plt.title("Flipkart Product Reviews Per Hour")
plt.xlabel("Hour")
plt.ylabel("Number of Reviews")

plt.xticks(hours)

plt.show()

```
