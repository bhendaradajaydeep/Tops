```python

import numpy as np

times = [30, 32, 28, 35, 40, 29, 31, 33, 36, 38,
         27, 45, 50, 34, 37, 39, 41, 42, 43, 44]

p25 = np.percentile(times, 25)
p50 = np.percentile(times, 50)
p75 = np.percentile(times, 75)

print("25th Percentile:", p25)
print("50th Percentile (Median):", p50)
print("75th Percentile:", p75)
```

```python
import numpy as np

scores = [10, 25, 30, 45, 50, 60, 70, 80, 90, 100]

Q1 = np.quantile(scores, 0.25)
Q2 = np.quantile(scores, 0.50)
Q3 = np.quantile(scores, 0.75)

print("Q1:", Q1)
print("Q2 (Median):", Q2)
print("Q3:", Q3)
```

```python
import numpy as np

steps = [6000,6500,7000,7200,6800,7500,8000,8200,7800,7700,
         7600,7400,7300,7100,6900,6700,6600,6400,6300,6200,
         8100,8300,8500,8700,8900,9100,9200,9400,9600,9800]

Q1 = np.quantile(steps, 0.25)
Q3 = np.quantile(steps, 0.75)

IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
```
