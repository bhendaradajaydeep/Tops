```python
from scipy.stats import ttest_1samp

# Sample delivery times (minutes)
delivery_times = [28, 32, 27, 35, 29, 31, 30, 33, 26, 34]

# Perform one-sample t-test
t_stat, p_value = ttest_1samp(delivery_times, 30)

print("T-Statistic =", round(t_stat, 4))
print("P-Value =", round(p_value, 4))

# Decision
if p_value < 0.05:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```

```python
from scipy.stats import ttest_ind

# Daily steps before installing the app
before = [4200, 4500, 3900, 5000, 4700, 4300, 4100, 4600, 4400, 4800]

# Daily steps after installing the app
after = [6500, 6800, 6200, 7100, 6900, 6700, 6400, 7200, 7000, 6800]

# Independent t-test
t_stat, p_value = ttest_ind(before, after)

print("T-Statistic =", round(t_stat, 4))
print("P-Value =", round(p_value, 4))

if p_value < 0.05:
    print("Reject the Null Hypothesis")
    print("Average daily steps increased significantly.")
else:
    print("Fail to Reject the Null Hypothesis")
```

```python
from scipy.stats import ttest_rel

# Screen time before Focus Mode
before = [6.5, 7.0, 5.8, 6.2, 7.1, 6.8, 5.9, 6.7, 6.3, 7.2]

# Screen time after Focus Mode
after = [5.2, 5.8, 4.9, 5.3, 6.0, 5.7, 4.8, 5.5, 5.2, 6.1]

# Paired t-test
t_stat, p_value = ttest_rel(before, after)

print("T-Statistic =", round(t_stat, 4))
print("P-Value =", round(p_value, 4))

if p_value < 0.05:
    print("Reject the Null Hypothesis")
    print("Focus Mode significantly reduced screen time.")
else:
    print("Fail to Reject the Null Hypothesis")
```

```python
from scipy.stats import ttest_ind

# Ratings of Zomato
zomato = [4.5, 4.2, 4.6, 4.1, 4.3, 4.4, 4.2, 4.7, 4.5, 4.4, 4.3, 4.6, 4.2, 4.5, 4.4]

# Ratings of Swiggy
swiggy = [4.1, 4.0, 4.2, 4.3, 4.0, 4.1, 4.2, 4.3, 4.1, 4.0, 4.2, 4.1, 4.3, 4.2, 4.0]

# Independent t-test
t_stat, p_value = ttest_ind(zomato, swiggy)

print("T-Statistic =", round(t_stat, 4))
print("P-Value =", round(p_value, 4))

if p_value < 0.05:
    print("Reject the Null Hypothesis")
    print("The average ratings are significantly different.")
else:
    print("Fail to Reject the Null Hypothesis")
```

```python
from scipy.stats import ttest_1samp

# Sample daily UPI transactions per user
transactions = [18, 20, 21, 19, 22, 20, 23, 19, 21, 20]

# Test against average of 20 transactions
t_stat, p_value = ttest_1samp(transactions, 20)

print("T-Statistic =", round(t_stat, 4))
print("P-Value =", round(p_value, 4))

if p_value < 0.05:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```
