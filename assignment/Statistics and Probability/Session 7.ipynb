```python
# Bayes Theorem Function
def bayes_churn_probability(prior_churn,
                            prob_email_given_churn,
                            prob_email):
    """
    Bayes' Theorem

    P(Churn | Email) =
    [P(Email | Churn) * P(Churn)] / P(Email)
    """

    return (prob_email_given_churn * prior_churn) / prob_email


result = bayes_churn_probability(
    prior_churn=0.15,
    prob_email_given_churn=0.6,
    prob_email=0.25
)

print("\nTask 3")
print("Probability of Churn given Email =", round(result, 4))
print("Percentage =", round(result * 100, 2), "%")

```
